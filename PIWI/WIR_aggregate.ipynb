### Site Details files - import and concatenate all sheets from all files

```python
all_SiteDetails = pd.DataFrame()
for f in glob.glob("SiteDetails/*.xlsx"):
    df = pd.read_excel(f, sheetname = 'Site Details')
    all_SiteDetails = all_SiteDetails.append(df)

all_Datums = pd.DataFrame()
for f in glob.glob("SiteDetails/*.xlsx"):
    df = pd.read_excel(f, sheetname = 'Datums')
    all_Datums = all_Datums.append(df)
    
all_Construction = pd.DataFrame()
for f in glob.glob("SiteDetails/*.xlsx"):
    df = pd.read_excel(f, sheetname = 'Construction')
    all_Construction = all_Construction.append(df)
    
all_Status = pd.DataFrame()
for f in glob.glob("SiteDetails/*.xlsx"):
    df = pd.read_excel(f, sheetname = 'Status')
    all_Status = all_Status.append(df)

all_LithologyLog = pd.DataFrame()
for f in glob.glob("SiteDetails/*.xlsx"):
    df = pd.read_excel(f, sheetname = 'Lithology Log')
    all_LithologyLog = all_LithologyLog.append(df)

all_SummaryLog = pd.DataFrame()
for f in glob.glob("SiteDetails/*.xlsx"):
    df = pd.read_excel(f, sheetname = 'Summary Log')
    all_SummaryLog = all_SummaryLog.append(df)

all_SiteAquifers = pd.DataFrame()
for f in glob.glob("SiteDetails/*.xlsx"):
    df = pd.read_excel(f, sheetname = 'Site Aquifers')
    all_SiteAquifers = all_SiteAquifers.append(df)
```
